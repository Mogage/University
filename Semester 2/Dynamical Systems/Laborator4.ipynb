```python
t=var('t')
k=var('k')
x0=var('x0')

x=function('x')(t)
deq1 = diff(x, t) == -k * x
ans = desolve(deq1, [x, t], ics=[0, x0])
sol(t, k, x0) = ans
show(sol) # ex 1 a

g = plot(sol(t, 1, 10), t, 0, 10, color="green")
g += plot(sol(t, 0.5, 10), t, 0, 10, color="red")
g += plot(sol(t, 0.1, 1), t, 0, 10, color="yellow")
g += plot(sol(t, 0.5, 1), t, 0, 10)
show(g) # ex 1 b
```

```python
t12=var('t12')

eq = sol(t12, k, x0) == x0/2
ans = solve(eq, k)
k1(t12)=ans[0].rhs()
kc14=k1(5730)
show(kc14) # ex 1 c
show(numerical_approx(kc14))
```
