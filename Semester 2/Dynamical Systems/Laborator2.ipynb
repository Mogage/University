```python
reset()
x = var('x')
y = function('y')(x)
eqd = diff(y,x) == (2*x)*(1+y^2)
eqd
```

```python
desolve(eqd, y)
```

```python
show(desolve(eqd, y))
```

```python
sol=desolve(eqd, y)
show(solve(sol, y))
```

```python
_C = var('_C')
sol1 = sol.substitute(_C == 1)
sol1 = solve(sol1, y)
sol1 = sol1[0].rhs()
plot(sol1, x, -1, 1)
```

```python
eqd = (x^2 - 1)* diff(y, x) + 2 * x * y ^ 2 == 0
sol = desolve(eqd, y)
show(sol)
```

```python
show(solve(sol, y))
```

```python
sol1 = sol.substitute(_C == 1)
sol1 = solve(sol1, y)
sol1 = sol1[0].rhs()
plot(sol1, x, 2, 10)
```

```python
eqd = diff(y, x, 2) - y == exp(2*x)
sol = desolve(eqd, y)
show(sol)
```

```python
_K1, _K2 = var('_K1, _K2')
sol1 = sol.substitute(_K1 == 1, _K2 == 3)
plot(sol1, x, -2, 2)
```

```python
eqd = diff(y,x) == 1 + y ^ 2
sol = desolve(eqd, y, ics=[0,1])
show(sol)
```

```python
sol1 = solve(sol, y)
sol1 = sol1[0].rhs()
plot(sol1, x, 0.7, 0.9)
```

```python
eqd = diff(y, x, 2) - 5 * diff(y, x) + 4*y == 0
sol = desolve(eqd, y, ics=[0, 5, 8])
show(sol)
```

```python
plot(sol, x, -1, 1)
```

```python
eqd = diff(y, x, 2) + 4 * y == 4*(sin(2*x) + cos(2*x))
sol = desolve(eqd, y, ics=[pi, 2*pi, 2*pi])
show(sol)
```

```python
plot(sol, x, 0, pi/2)
```

```python
k = var('k')
eqd = diff(y, x) + k/x * y == x ^ 3
sol = desolve(eqd, y, ivar=x)
show(sol)
```

```python
sol1 = sol.substitute(k=1)
sol1 = sol1.substitute(_C=1)
show(sol1)
```

```python
plot(sol1, x, -1, 1, ymin=-10, ymax=10)
```

```python
sol = desolve(eqd, y, ivar=x, ics=[1, 0])
show(sol)
```

```python
sol1 = sol.substitute(k=1)
plot(sol1, x, -1, 1, ymin=-10, ymax=10)
```
